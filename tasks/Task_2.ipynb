# Задание 2.14 (8 баллов)

Постройте график функции $$f(x) = (5  \sin(2x) - \cos^2(x) - 1 + |x^3| - x^2) / (x^2 + 1)$$  

и производной от этой функции.  Найдите минимум с помощью `scipy` и обозначьте его на графике. На этой же оси постройте графики скользящего среднего этой функции с окном $1$, $2$, $5$.

```python
import math
from scipy.misc import derivative
from scipy import optimize

def moving_average(a, n):
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n

x = np.arange(-10, 10, 0.01)

def function(x):
    return (5*np.sin(2*x) - pow(np.cos(x), 2) - 1 + np.abs(pow(x, 3)) - pow(x, 2))/(pow(x, 2) + 1)
plt.plot(x, function(x), color="purple", alpha = 0.3)

derivative_function = derivative(function, x, dx=1e-9)
plt.plot(x, derivative_function, color="green")



func_min = optimize.minimize(function, x0=0, method="BFGS").fun
x_min = optimize.minimize(function, x0=0, method="BFGS").x[0]
plt.plot(x_min, func_min, 'ro')



window1_F = moving_average(function(x), 1)
args_window1_F = moving_average(x, 1)
window2_F = moving_average(function(x), 2)
args_window2_F = moving_average(x, 2)
window5_F = moving_average(function(x), 5)
args_window5_F = moving_average(x, 5)

plt.plot(args_window1_F, window1_F, color="orange", alpha=1)
plt.plot(args_window2_F, window2_F, color="purple", alpha=1)
plt.plot(args_window5_F, window5_F, color="pink", alpha=1)


plt.show()
```

# Задание 2.15 (10 баллов)

Есть следующий алгоритм семплинирования, на вход функции подается вектор из целых положительных чисел и число семплов.

```python
from typing import List 
from scipy.stats import ttest_ind

def sample(x: List[int], c: int) -> List[int]:
    assert len(x) > 0
    
    s = np.sum(x) 
    # print(s, np.random.random())
    res = []
    for _ in range(c):
        
        val = s * np.random.random()
        cur, idx = 0, 0        
        while cur + x[idx] <= val: # пока текущее значение + порядковый элемент (от 0) массива не больше суммы всех элементов * rand(0,1)
            cur += x[idx] #увеличиваем текущее значение НА значение используемого порядкого элемента
            idx += 1 #переходим к следующему элементу
            # в итоге у нас добавляется индекс (итерация), на котором текущие условия выполнились
            # по сути если rand(0,1) будет маленьким каждый раз, мы можем получить 0,0,0... (c раз)
        res.append(idx)
    return res  

default_sample = sample([50, 3, 1, 7, 20], 500) 

def numpy_sample(x: List[int], c: int) -> List[int]:
    assert len(x) > 0

    x_ = np.array(x)
    cumsum_x_ = x_.cumsum()

    return [np.argmax(cumsum_x_ > np.sum(x_) * np.random.random()) for _ in range(c)]

n_sample = numpy_sample([50, 3, 1, 7, 20], 500)

# effect_value = np.mean(default_sample) - np.mean(n_sample) / np.std(default_sample)
ttest_ind(default_sample, n_sample, alternative="two-sided").pvalue
# Можем заметить, что значение T - статистики практически всегда > 0.05, значит нет основания отвергать
# нулевую гипотезу об однородности выборок N_sample (с numpy) и default_sample, значит принимается нулевая
# гипотеза

```

Есть следующий алгоритм "случайного блуждания"

```python
import random 
from time import sleep
from seaborn import kdeplot

%timeit
def walk(n: int):
    cur_position = 0
    for i in range(n):
        cur_position += random.choice([-3, -1, 0, 2, 3])

    return cur_position

%time walk(10_000)
```
